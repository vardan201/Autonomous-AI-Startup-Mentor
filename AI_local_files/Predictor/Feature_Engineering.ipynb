```python
import pandas as pd

# Convert founded_at and last_funding_at to year
data["founded_year"] = pd.to_datetime(
    data["founded_at"], errors="coerce"
).dt.year

data["last_funding_year"] = pd.to_datetime(
    data["last_funding_at"], errors="coerce"
).dt.year

# Compute startup age
data["startup_age"] = data["last_funding_year"] - data["founded_year"]

# Ensure no negative ages
data["startup_age"] = data["startup_age"].clip(lower=0)
```

```python
DROP_COLS = [
    # Junk / unnamed columns
    "Unnamed: 0", "Unnamed: 6",

    # Identifiers / names
    "id", "object_id", "name", "labels",

    # Location-specific (US bias)
    "city", "state_code", "state_code.1", "zip_code",
    "latitude", "longitude",
    "is_CA", "is_NY", "is_MA", "is_TX", "is_otherstate",

    # Raw date columns (we already used them)
    "founded_at", "closed_at",
    "first_funding_at", "last_funding_at",

    # Industry one-hot columns (too granular, US-centric)
    "category_code",
    "is_software", "is_web", "is_mobile", "is_enterprise",
    "is_advertising", "is_gamesvideo", "is_ecommerce",
    "is_biotech", "is_consulting", "is_othercategory",

    # Funding round flags (high leakage / redundant)
    "has_roundA", "has_roundB", "has_roundC", "has_roundD",

    # Age columns we don’t need anymore
    "age_first_funding_year",
    "age_first_milestone_year",
    "age_last_milestone_year",

    # Ranking (US-only signal)
    "is_top500"
]

# Drop columns in-place (safe)
data.drop(columns=[c for c in DROP_COLS if c in data.columns], inplace=True)
```

```python
data["startup_age_safe"] = data["startup_age"] + 1
data["funding_per_year"] = (
    data["funding_total_usd"] / data["startup_age_safe"]
)
```

```python
data["execution_velocity"] = (
    data["milestones"] / data["startup_age_safe"]
)
```

```python
data["rounds_per_year"] = (
    data["funding_rounds"] / data["startup_age_safe"]
)
```

```python
data.drop(columns=["startup_age_safe"], inplace=True)
```
